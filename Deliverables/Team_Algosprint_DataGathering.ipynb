# Handle Null Values. We cannot use 0 to fill the null values because non of the values can be zero in reality

```python
infrastructure_data.info()
```

```python
def get_rows_with_missing_data(dataframe):
    # Find rows with missing data
    rows_with_missing_data = dataframe[dataframe.isnull().any(axis=1)]
    return rows_with_missing_data
rows_missing_data = get_rows_with_missing_data(infrastructure_data)
rows_missing_data
```

```python
infrastructure_data.info()
```

```python


def interpolate_by_country(df):
    # Grouping the DataFrame by 'Country' column
    grouped = df.groupby('Country')
    
    # Interpolate the grouped data by country
    interpolated = grouped.apply(lambda group: group.interpolate(method='linear', limit_direction='both'))
    
    return interpolated.reset_index(drop=True)

# Interpolate missing values based on country
infrastructure_data = interpolate_by_country(infrastructure_data)

infrastructure_data.sort_values(by=["gdp"]).tail(50)
```

```python
def fill_null_with_random(df):
    for column in df.columns:
        if df[column].isnull().sum() > 0:  # Check for columns with null values
            random_values = np.random.uniform(277.94,102.21, size=df[column].isnull().sum())  # Generate random values
            df[column] = df[column].fillna(pd.Series(random_values, index=df[column][df[column].isnull()].index))  # Fill null values with random values

fill_null_with_random(infrastructure_data)
```
